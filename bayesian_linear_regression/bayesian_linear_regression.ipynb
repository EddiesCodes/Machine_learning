```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
# create X and y of the prior and likelihood
mean = np.array([1., 2.])
cov = np.array([[1., 0.9], [0.9, 1.]])
x, y = np.random.multivariate_normal(mean, cov, 100).T
X = np.stack([np.ones_like(x), x], axis=1)
print(X)
```

```python
#calculate the posterior
pos_cov = np.linalg.inv(X.T@X + np.eye(len(X.T@X)))
pos_mean = pos_cov@(X.T@y)
# posterior = N(w| pos_mean, pos_cov)
```

```python
#100 samples from the posterior
w = np.random.multivariate_normal(pos_mean, pos_cov, 100)
```

```python

plt.figure()
for i in range(100):
    plt.plot(x, X@w[i], 'b', lineWidth = 0.5)
plt.plot(x, y, 'rx');
```
